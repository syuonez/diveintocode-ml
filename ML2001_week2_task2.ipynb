# 【問題1】乱数の作成
平均が(-3, 0)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。


```python
from numpy.random import *
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
 
mu = [-3, 0]
sigma = [[1.0, 0.8], [0.8, 1]]

# 2次元正規乱数を500個生成
np.random.seed(0) 
values = multivariate_normal(mu, sigma, 500)
```

# 【問題2】散布図による可視化
問題1で作成したデータ点を散布図により可視化してください。

散布図はmatplotlibのplt.scatter()を使うことで描けます。

```python
plt.title("Scatter",fontsize=20) # タイトル
plt.xlabel("x1", fontsize=20) # x軸ラベル
plt.ylabel("x2", fontsize=20) # y軸ラベル

 
plt.scatter(values[:,0], values[:,1])
plt.show()
```

# 【問題3】ヒストグラムによる可視化
問題1で作成したデータをヒストグラムにより可視化してください。
ヒストグラムはplt.hist()を使うことで描けます。

各次元ごとに作成し、2つのヒストグラムとしてください。
それぞれのヒストグラムの横軸の範囲はplt.xlim()を用いて揃えるようにしてください。

```python
plt.hist(values[:,0], bins = 500)
plt.xlim([-6,3])
plt.title("histogram of x1",fontsize=20) # タイトル
plt.xlabel("x1", fontsize=20) # x軸ラベル
plt.ylabel("frequency", fontsize=20) # y軸ラベル

plt.show()
```

```python
plt.hist(values[:,1], bins = 500)
plt.xlim([-6, 3])
plt.title("histogram of x2",fontsize=20) # タイトル
plt.xlabel("x2", fontsize=20) # x軸ラベル
plt.ylabel("frequency", fontsize=20) # y軸ラベル

plt.show()
```

# 【問題4】データの追加
新たに平均が(0, -3)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。
そして、問題1、4それぞれのデータをひとつの散布図として可視化してください。凡例として問題1のものは0、問題2のものは1を表示してください。


```python
mu1 = [0, -3]
sigma1 = [[1.0, 0.8], [0.8, 1]]

# 2次元正規乱数を500個生成
np.random.seed(0) 
values1 = multivariate_normal(mu1, sigma1, 500)


plt.title("Scatter",fontsize=20) # タイトル
plt.xlabel("x1", fontsize=20) # x軸ラベル
plt.ylabel("x2", fontsize=20) # y軸ラベル


 
plt.scatter(values[:,0], values[:,1], label ='0')
plt.scatter(values1[:,0], values1[:,1], label = '1')
plt.legend()

plt.show()
```

# 【問題5】データの結合
データはまとめておいた方が後々扱いやすいです。問題1、4で作成したndarrayを 結合 し、(1000, 2)のndarrayとしてください。結合はnp.concatenate()やnp.vstack()を使うことで行えます。

```python
x = np.concatenate([values, values1], 0) #結合 縦
print(x.shape) #確認
```

# 【問題6】ラベル付け
ひとまとめになった1000個のデータそれぞれに対して、問題1、4どちらで作成したものなのかを示す ラベル 付けを行ってください。問題1のものには0、問題4のものには1を対応させます。

0と1を含むラベルの列を新たに追加し、(1000, 3)のndarrayを作成してください。

機械学習に使用するデータセットはこのような形になっていることが多いです。

```python
npdata_zero = np.zeros((500, 1),int) #(500,1)のndarray 作成 [0]
npdata_one = np.ones((500,1),int) #(500,1)のndarray 作成 [1]
npdata_rabel = np.concatenate([npdata_zero, npdata_one], 0) #縦に結合
npdata = np.concatenate([x, npdata_rabel], 1) # 横に結合
print(npdata)
print(npdata.shape) #確認(1000, 3)

print(npdata[496:503]) # データ497個目〜503個目のラベル確認　 501個目から０から１にラベルが変わっている
```
