# 【問題5】Pythonの関数化

問題1から問題3までの範囲をまとめて行うPythonの関数を雛形を参考に作成してください。
そして、以下の3式の勾配を0.1間隔で求めてndarrayで出力してください。

《雛形》

```python
def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    """
    変化量を使い勾配を計算する。
    
    Parameters
    ----------------
    function : function
    求めたい関数、xのndarrayを入れるとyのndarrayを返すもの。
    x_range : tuple
      np.arange()の引数と同じ方法で範囲を指定する。

    Returns
    ----------------
    array_xy : ndarray, shape(n, 2)
      xとyを結合したもの。nはx_rangeに依存する。
    gradient : ndarray, shape(n-1,)
      関数の勾配。差をとるため要素は1つ減る
    """
    x = np.arange(*x_range)
    array_xy = np.stack([x,function]).T
   # print(array_xy)
    
    gradient = (array_xy[1:,1] - array_xy[:-1,1]) / (array_xy[1:,0] - array_xy[:-1,0])
    return array_xy, gradient
```

```python
def function1(array_x):
    array_y = array_x**2
    return array_y
```

```python
def function2(array_x):
    array_y = (2*(array_x)**2)+(2**(array_x))
    return array_y
```

```python
def function3(array_x):
    array_y = np.sin(array_x.astype(np.complex) ** 0.5)
    return array_y
```

```python
a1 = function1(array_x) #y1
xy1, g1 = compute_gradient(a1)  #return array_xy, gradient  
print(xy1)
print(g1)
```

```python
a2 = function2(array_x) #y2
xy2, g2 = compute_gradient(a2)  #return array_xy, gradient  
print(xy2)  
print(g2)
```

```python
a3 = function3(array_x) #y3
xy3, g3= compute_gradient(a3)  #return array_xy, gradient  
print(xy3)  
print(g3)
```
